## Data Pre-Processing

```python
print("no of columns :",df.shape[1])
print("no of rows :",df.shape[0])
```

```python
#data info
df.info()
```

```python
# Check Duplicate data
df.duplicated()
```

```python
# check missing value
df.isnull().sum()
```

```python
# check missing value with persentase
def chek_missing(df):
    sum_nan = df.isna().sum().reset_index()
    sum_nan.columns = ['kolom', 'jumlah_NaN']
    sum_nan["persentase"] = sum_nan.jumlah_NaN/len(df)*100
    return sum_nan.sort_values("persentase", ascending =  False)
df_missing = chek_missing(df)
df_missing
```

```python
#missing value comparison
plt.figure(figsize=(12,20))
sns.barplot(y = df_missing['kolom'], x = df_missing['persentase']).set_title("Missing Value Comparison");
```

```python
#handling missing value
df['cast'].fillna('unknown', inplace = True)
#df['country'].fillna('unknown', inplace = True)
df.dropna(subset=['date_added','rating','duration','country'], inplace = True)
```

```python
df = df.copy()
df = df.fillna('NULL')
```

```python
#top 10 netflix movie and tv show by duration
df['duration'].value_counts(ascending = False).head(10).plot(kind = 'bar', color = 'green')
plt.xticks(rotation = 45)
plt.show()
```

```python
#counting the number of film and tv show productions from various countries
country_type = df[df['country'] != "NULL"].groupby(by = ['country','type'])['country'].count().to_frame()
country_type.columns = ['count']
country_type.reset_index()
country_type = country_type.sort_values(by=['count','country'] ,  ascending = False).head(15).reset_index()
plt.figure(figsize = [20,10]);
sns.barplot(data=  country_type , x = 'country' , y = 'count' , hue = 'type');
plt.xticks(rotation = 90);
```

```python
#Count the number of movies and tv shows by release year
release_year = df[df['release_year'] != "NULL"].groupby(by = ['release_year','type'])['release_year'].count().to_frame()
release_year.columns = ['count']
release_year.reset_index()
release_year = release_year.sort_values(by=['count','release_year'] ,  ascending = False).head(10).reset_index()
plt.figure(figsize = [20,10]);
sns.barplot(data=  release_year , x = 'release_year' , y = 'count' , hue = 'type');
plt.xticks(rotation = 90);
```
